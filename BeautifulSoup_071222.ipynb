# QUIZ 2

```python
#send request
page1=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")

page1
```

```python
#scraping the content
bsoup=BeautifulSoup(page1.content)
```

```python
#sraping the movies name
movies_name=[]
for i in bsoup.find_all('h3',class_="lister-item-header"):
    movies_name.append(i.text.replace("\n"," ").split("(")[0])
movies_name
```

```python
#scraping the year of movies
year=[]
for i in bsoup.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text.replace("("," ").split(")")[0])
year
```

```python
#scraping the rating of movies
rating=[]
for i in bsoup.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace("\n"," "))
rating
```

```python
#importing the libraries
import pandas as pd
```

```python
#for checking the Length
print(len(movies_name),len(year),len(rating))
```

```python
#to make the data frame
Df= pd.DataFrame({"Name":movies_name,"Year":year,"Rating":rating})

Df
```

```python
page2=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
page2
```

```python
bsoup=BeautifulSoup(page2.content)
```

```python
#scraping the bollywood movies name
movies=[]
for i in bsoup("td",class_="titleColumn"):
    movies.append(i.text.replace("\n"," ").split("(")[0])
movies
```

```python
#scraping the year of movies 
Year=[]
for i in bsoup("span",class_="secondaryInfo"):
    Year.append(i.text.replace("("," ").split(")")[0])
Year
```

```python
#rating of movies name
rating=[]
for i in bsoup("td",class_="ratingColumn imdbRating"):
    rating.append(i.text.replace("\n"," "))
rating
```

```python
#check the length of all above data
print(len(movies),len(Year),len(rating))
```

```python
#import pandas lirbray
import pandas as pd
```

```python
#to make the data frame
df=pd.DataFrame({"Movies_name":movies,"Year":Year,"Rating":rating})
df
```
